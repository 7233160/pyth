```python
import pandas as pd
import json
import csv
```

```python
li=[]
with open("data.json") as datasrc:
    for line in datasrc:
        jline=json.loads(line)
        li.append(jline['data']['Author'])
```

```python
li
```

```python
import re
```

```python
li1=[]
for item1 in li:
    regex=re.compile(r'(\w)+(\s)?(\w)+')
    li1.append(regex.search(item1).group())
```

```python
li1
```

```python
b={}
from collections import Counter
b=dict(Counter(li1))
```

```python
b
```

```python
namelist=[]
commitlist=[]
for k,v in b.items():
    namelist.append(k)
    commitlist.append(v)
```

```python
print(namelist)
```

```python
print(commitlist)
```

```python
dict1={}
dict1['Name']=namelist
dict1['Commit']=commitlist
```

```python
dict1
```

```python
data=pd.DataFrame(dict1)
```

```python
data
```

```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
top_10=data.sort_values('Commit',ascending=False).head(10)
```

```python
top_10
```

```python
plt.figure(figsize=(12,6))
x=range(10)
plt.bar(x,top_10['Commit']/10)
plt.xticks(x,top_10['Name'])
plt.xlabel('Name')
plt.ylabel('Number Of Commits')
plt.title('Top 10 Contributors')
plt.show()
```
